```python
from random import choice
from time import sleep
```

```python
def chk( grid ) :
    for i in grid :
        if i[0] == i[1] == i[2] :
            return i[0]
    for i in range(3) :
        if grid[0][i] == grid[1][i] == grid[2][i] :
            return grid[0][i]
    if grid[0][0] == grid[1][1] == grid[2][2] :
        return grid[0][0]
    if grid[0][2] == grid[1][1] == grid[2][0] :
        return grid[1][1]
    return 0
```

```python
def printGrid( grid ) :
    print()
    for l in grid :
        for i in l :
            print ( i,end=" " )
        print()
    print()
```

```python
def mode2() :
    grid = [ [ 3*y + i+1 for i in range(3)] for y in range(3) ]
    cpu_domain = [ i+1 for i in range(9) ]
    
    p1n = input("\nPlayer Name : \t")
    p1c = input("Player Choice O\X : \t")
    
    if p1c == 'X' : p2c = 'O'
    else : p2c = 'X'
    
    turn = 0 #turn 1 means plyr2 and opp
    while True :
        printGrid(grid)
        pos = -1
        if turn :
            print ("Turn of Computer",p2c)
            sleep(1)
            pos = choice(cpu_domain)
            print ("Computer chose :",pos)
            sleep(1)
        else :
            print ("Turn of",p1n,p1c)
            pos = int(input("Enter position:            "))
        x = (pos-1)//3
        y = (pos-1)%3
        if str(grid[x][y]) in "XO" :
            print("Invalid Move !\nTry Again")
        else :
            if turn==1 :
                grid[x][y] = p2c
            else :
                grid[x][y] = p1c
            del cpu_domain[cpu_domain.index(pos)]
            turn+=1
            turn%=2
        
        res = chk(grid)
        if res==0 :
            continue
        
        if res == p2c :
            print("\nShah! Compuer Won !!\n")
            printGrid(grid)
            break 
        else :
            print("\nYeah!\nCongrats",p1n,"you won !!\n")
            printGrid(grid)
            break 
        
```
